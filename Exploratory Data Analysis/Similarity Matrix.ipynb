# Exploratory Data Analysis

## Import Libraries

```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

## Loading the Dataset

```python
file_path = '../Data/Product Flags/cust_info_prod_202403_processed.parquet'
march_2024 = pd.read_parquet(file_path)
march_2024
```

```python


# Assuming 'march_2024' is your DataFrame, and it includes only product-related columns.
# Select only the product usage columns (exclude CUST_NUM or any other non-relevant columns)
product_columns = ['CASH_MANAGEMENT', 'CORPORATE_LOANS', 'RETAIL_LOAN', 'TRADE_SERVICES', 
                   'INVESTMENT_FUNDS', 'SECURITIES', 'INSURANCE', 'CORPORATE_FINANCE', 
                   'FOREX', 'REMITTANCE', 'CORPORATE_CARDS', 'DEPOSITS', 
                   'RETAIL_PRODUCTS', 'INVESTMENTS_AND_SECURITIES']

# Filter the dataset to only include these columns
product_data = march_2024[product_columns]

# Normalize the data (optional, if needed) to scale product usage between 0 and 1
# product_data = product_data / product_data.max()

### Step 2: Calculate Cosine Similarity ###
# Calculate the cosine similarity between customers
similarity_matrix = cosine_similarity(product_data)

# Convert the similarity matrix to a DataFrame for better readability
similarity_df = pd.DataFrame(similarity_matrix, index=march_2024['CUST_NUM'], columns=march_2024['CUST_NUM'])

# Display the similarity matrix
similarity_df.head()
```
